## (1) 모든 프로젝트에 들어가는 공통 모듈 import 하기¶

```python
# scikit-learn 라이브러리 임포트
from sklearn.model_selection import train_test_split #데이터셋 분리 함수
from sklearn.metrics import classification_report #분류 성능평가 함수

from sklearn.tree import DecisionTreeClassifier #의사결정나무 함수
from sklearn.metrics import accuracy_score #정확도 확인함수

from sklearn.ensemble import RandomForestClassifier #RandomForest 함수

from sklearn import svm #svm 함수

from sklearn.linear_model import SGDClassifier #SGDClassifier 함수

from sklearn.linear_model import LogisticRegression #LogisticRegression 함수

# pandas 라이브러리 임포트
#행과 열이 있는 2차원 데이터를 다룰때 사용. 표 데이터 분석이나, 대형 데이터의 여러 통계량 다룰때 사용 
import pandas as pd

print('import 완료!!')
```

## load_breast_cancer (2) breast_cancer을 import 하고, 데이터 준비하기

```python
from sklearn.datasets import load_breast_cancer #breast_cancer 데이터 로딩하기

breast_cancer = load_breast_cancer()

print(dir(breast_cancer))
```

## load_breast_cancer (3) 데이터 이해하기

```python
#Feature data 지정하기
breast_cancer_data=breast_cancer.data
print(breast_cancer_data.shape)
breast_cancer_data
```

```python
#Label data 지정하기
breast_cancer_label = breast_cancer.target
print(breast_cancer_label.shape)
breast_cancer_label
```

```python
#Target Names 출력해보기
breast_cancer.target_names
```

```python
#데이터 Describe 해 보기
print(breast_cancer.DESCR)
```

## load_breast_cancer (4) train, test 데이터 분리

### breast_cancer을 pandas가 제공하는 DataFrame(자료형) 으로 변환하기

```python
breast_cancer_df = pd.DataFrame(data=breast_cancer_data, columns=breast_cancer.feature_names)
breast_cancer_df["label"] = breast_cancer.target
breast_cancer_df
```

### test dataset과 training dataset으로 분리하기

```python
X_train, X_test, y_train, y_test = train_test_split(breast_cancer_data, 
                                                    breast_cancer_label, 
                                                    test_size=0.2, 
                                                    random_state=8)

print('X_train 개수: ', len(X_train),', X_test 개수: ', len(X_test))
```

## load_breast_cancer (5) 다양한 모델로 학습시켜보기

```python
#Decision Tree 사용해 보기

#train 학습
decision_tree = DecisionTreeClassifier(random_state=8)
decision_tree.fit(X_train, y_train)

#test 예측
y_pred_DT_breast_cancer = decision_tree.predict(X_test)

#정확도 확인
acc_DT_breast_cancer= accuracy_score(y_test, y_pred_DT_breast_cancer)
acc_DT_breast_cancer
```

```python
#Random Forest 사용해 보기

#train 학습
random_forest = RandomForestClassifier(random_state=8)
random_forest.fit(X_train, y_train)

#test 예측
y_pred_RF_breast_cancer = random_forest.predict(X_test)

#정확도 확인
acc_RF_breast_cancer = accuracy_score(y_test, y_pred_RF_breast_cancer)
acc_RF_breast_cancer
```

```python
#SVM 사용해 보기

#train 학습
svm_model = svm.SVC(random_state=8)
svm_model.fit(X_train, y_train)

#test 예측
y_pred_SVC_breast_cancer = svm_model.predict(X_test)

#정확도 확인
acc_SVC_breast_cancer = accuracy_score(y_test, y_pred_SVC_breast_cancer)
acc_SVC_breast_cancer
```

```python
#SGD Classifier 사용해 보기

#train 학습
sgd_model = SGDClassifier(random_state=8)

#test 예측
sgd_model.fit(X_train, y_train)
y_pred_SGD_breast_cancer = sgd_model.predict(X_test)

#정확도 확인
acc_SGD_breast_cancer = accuracy_score(y_test, y_pred_SGD_breast_cancer)
acc_SGD_breast_cancer
```

## load_breast_cancer (6) 모델을 평가해 보기

```python
print("Decision Tree")
print(classification_report(y_test, y_pred_DT_breast_cancer))
print()

print("Random Forest")
print(classification_report(y_test, y_pred_RF_breast_cancer))
print()

print("SVM")
print(classification_report(y_test, y_pred_SVC_breast_cancer))
print()

print("SGD Classifier")
print(classification_report(y_test, y_pred_SGD_breast_cancer))
```
